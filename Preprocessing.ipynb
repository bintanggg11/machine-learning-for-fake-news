```python
import re

def hapus_selengkapnya(teks):
    if isinstance(teks, str):
        teks = re.sub(r'Simak (selengkapnya|informasinya) dalam video berikut.*', '', teks)
        teks = teks.replace("KOMPAS.com - ", "")
    return teks

# Terapkan fungsi pada dataframe
df_kompas['FullText'] = df_kompas['FullText'].apply(hapus_selengkapnya)
```

```python
df_kompas['FullText'].iloc[9]
```

```python
def wardrop(text):

    text = text.lower()
    return text

df_kompas['FullText'] = df_kompas['FullText'].apply(wardrop)
```

```python
# Import library yang diperlukan
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
import pandas as pd

# Membuat instance StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

df_kompas.at[18, 'FullText'] = stopword.remove(df_kompas.at[18, 'FullText'])
```

```python
# Import library yang diperlukan
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat instance StopWordRemoverFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

df_kompas.at[18, 'FullText'] = stemmer.stem(df_kompas.at[18, 'FullText'])
```

```python
# Fungsi untuk menghapus spasi berlebihan menggunakan regex
def hapus_spasi_berlebihan(teks):
    return re.sub(r'\s+', ' ', teks).strip()

# Terapkan fungsi pada dataframe
df_kompas['FullText'] = df_kompas['FullText'].apply(hapus_spasi_berlebihan)
```

# CNN

```python
df_cnn = pd.read_excel('Data\\CNN.xlsx')
df_cnn
```

```python
df_cnn = df_cnn[['FullText', 'hoax']]
```

```python
df_cnn['FullText'].iloc[1]
```

```python
def hapus_lihat_juga(teks):
    return re.sub(r'\bLihat Juga\b.*', '', teks)

df_cnn['FullText'] = df_cnn['FullText'].apply(hapus_lihat_juga)
```

```python
def baris_baru(teks):
    return ' '.join(teks.strip().split('\n'))

df_cnn['FullText'] = df_cnn['FullText'].apply(baris_baru)
```

```python
def hapus_kalimat_terakhir(teks):
    pattern = r'\([^)]+\)$'  # Pola untuk menemukan kalimat terakhir dalam tanda kurung
    teks_baru = re.sub(pattern, '', teks)
    return teks_baru.strip()  # Menghapus spasi tambahan di awal atau akhir teks

df_cnn['FullText'] = df_cnn['FullText'].apply(hapus_kalimat_terakhir)
```

```python
def remove_extra_spaces(df):
    df = df.applymap(lambda x: ' '.join(x.split()) if isinstance(x, str) else x)
    return df

df_cnn = remove_extra_spaces(df_cnn)
```

```python
def extract_explanation(text):
    # Pisahkan teks berdasarkan string "[PENJELASAN]:"
    explanation_split = text.split("[PENJELASAN]:")
    explanation_split2 = text.split("Penjelasan:")
    explanation_split3 = text.split("PENJELASAN:")
    explanation_split4 = text.split("PENJELASAN :")
    explanation_split5 = text.split("Penjelasan :")
    explanation_split6 = text.split("[PENJELASAN]")

    # Jika terdapat penjelasan, ambil kalimat penjelasannya saja
    if len(explanation_split) > 1:
        explanation = explanation_split[1].strip()
        return explanation
    elif len(explanation_split2) > 1:
        explanation = explanation_split2[1].strip()
        return explanation
    elif len(explanation_split3) > 1:
        explanation = explanation_split3[1].strip()
        return explanation
    elif len(explanation_split4) > 1:
        explanation = explanation_split4[1].strip()
        return explanation
    elif len(explanation_split5) > 1:
        explanation = explanation_split5[1].strip()
        return explanation
    elif len(explanation_split6) > 1:
        explanation = explanation_split6[1].strip()
        return explanation
    elif len(explanation_split6) > 0:
        explanation = explanation_split6[0].strip()
        return explanation
    else:
        return "Penjelasan tidak ditemukan."

df_hoax['FullText'] = df_hoax['FullText'].apply(extract_explanation)
```

```python
import re

def hapus_teks(teks):
    # Menggunakan regex untuk mencari pola "="
    pattern = re.compile(r'=+')
    match = pattern.search(teks)
    if match:
        # Menghapus teks setelah kecocokan pertama
        return teks[:match.start()].strip()
    else:
        return teks

# Terapkan fungsi ke kolom 'text'
df_hoax['FullText'] = df_hoax['FullText'].apply(hapus_teks)
```

```python
import re

def hapus_kata_kunci(teks):
    # Menggunakan regex untuk menghapus kalimat dengan kata 'https:' dan 'www.'
    teks_baru = re.sub(r'https?:\S+|www\.\S+', '', teks)
    return teks_baru.strip()

df_hoax['FullText'] = df_hoax['FullText'].apply(hapus_kata_kunci)
```
